```
img1 = cv2.imread('/content/drive/My Drive/cv2_test/1.jpg')
img2 = cv2.imread('/content/drive/My Drive/cv2_test/2.jpg')
img3 = cv2.imread('/content/drive/My Drive/cv2_test/3.jpg')
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(img1)
cv2_imshow(img2)
cv2_imshow(img3)
```

```
  hsv1 = cv2.cvtColor(img1,cv2.COLOR_BGR2HSV)
  hsv2 = cv2.cvtColor(img2,cv2.COLOR_BGR2HSV)
  hsv3 = cv2.cvtColor(img3,cv2.COLOR_BGR2HSV)
```

```
cv2_imshow(hsv1)
cv2_imshow(hsv2)
cv2_imshow(hsv3)
```

```
low_range = np.array([0,155,155])
up_range = np.array([60,255,255])
mask1 = cv2.inRange(hsv1, low_range , up_range)
mask2 = cv2.inRange(hsv2, low_range , up_range)
mask3 = cv2.inRange(hsv3, low_range , up_range)
cv2_imshow(mask1)
cv2_imshow(mask2)
cv2_imshow(mask3)
```

```
edges1 =  cv2.Canny(mask1 , 0 , 600)
edges2 =  cv2.Canny(mask2 , 0 , 600)
edges3 =  cv2.Canny(mask3 , 0 , 600)
cv2_imshow(edges1)
cv2_imshow(edges2)
cv2_imshow(edges3)
```
