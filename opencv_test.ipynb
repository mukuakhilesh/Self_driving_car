```
img = cv2.imread('/content/drive/My Drive/cv2_test/1.jpg')
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```
gray = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```
cv2_imshow(gray)
```

```
low_range = np.array([0,155,155])
up_range = np.array([60,255,255])
mask = cv2.inRange(gray , low_range , up_range)
cv2_imshow(mask)
```

```
edges =  cv2.Canny(mask , 0 , 600)
cv2_imshow(edges)
```
